```python
import librosa
import librosa.display
import IPython.display
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import pandas as pd
import scipy as sklearn
import os
```

```python
filename = np.array(pd.read_csv('../MFCC+DELTA/SC_file_name.csv'))
answer = np.array(pd.read_csv('../MFCC+DELTA/SC_answer.csv'))

for step in range(188):
    audio_path=audio_path='../MFCC+DELTA/SC/'+filename[step]
    
    def MFCC(y,sr,mfccs):
            y, sr = librosa.load(audio_path[0]) 
            mfccs=librosa.feature.mfcc(y=y,sr=sr)
            plt.figure(figsize=(0.78*size_,0.78*size_)) #size 56x56
            librosa.display.specshow(mfccs)
            plt.title(audio_path[0])
            plt.tight_layout()
            
print(MFCC)
```
